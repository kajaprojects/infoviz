```python
import pandas as pd
import numpy as np
import panel as pn
import plotly.express as px
pn.extension('tabulator')
from dash import Dash, dcc, html, Input, Output
import plotly.graph_objects as go
import panel.widgets as pnw
import hvplot.pandas
```

```python
#df = pd.read_csv('C:\\Users\\Kaja\\Desktop\\infoviz\\data\\olympic_medals.csv')
df = pd.read_csv('C:\\Users\\Kaja\\Desktop\\infoviz\\data\\olympic_results.csv')
```

```python
df = df.fillna(0)
idf = df.interactive()
```

```python
#getting a date from slug_game
df['date'] = df['slug_game'].str.split('-').str[-1]
df['date'] = df['date'].str.strip()
#df
```

```python
#group by types of medals per year
df2 = df
worldmap1 = df.groupby(['date','country_name', 'medal_type']).size().reset_index(name='counts')
worldmap1 = worldmap1[worldmap1['medal_type'] != 0] 
#vis3 dataframe
disc_df = df.groupby(['date', 'discipline_title','country_name', 'medal_type']).size().reset_index(name='counts')
disc_dfx = disc_df[disc_df['medal_type'] != 0] 

#new df for second vis so it's not slow
disc_df2 = df.groupby(['date', 'discipline_title','country_name', 'medal_type']).size().reset_index(name='counts')
disc_dfx2 = disc_df[disc_df['medal_type'] != 0] 
#worldmap.head
worldmapx = worldmap1[worldmap1['medal_type'] != 0]
# df with sum of all medals by country and date
worldmap = worldmap1.groupby(['date','country_name']).size().reset_index(name='sum_of_medals')
#worldmap['gold'] = worldmap1['medal_type'] == 'GOLD'

worldmap

#df for slovenia
slovenia_df = worldmapx[worldmapx['country_name'] == 'Slovenia']
slovenia_df

#slovenia_df['gold_medals'] = slovenia_df['counts'].where(['medal_type']=='GOLD', slovenia_df['medal_type'])
slovenia_df.loc[slovenia_df['medal_type'] == 'GOLD', 'gold_medal'] = slovenia_df['counts']
slovenia_df['gold_medal'].fillna(0, inplace=True)
slovenia_df.loc[slovenia_df['medal_type'] == 'SILVER', 'silver_medal'] = slovenia_df['counts']
slovenia_df['silver_medal'].fillna(0, inplace=True)
slovenia_df.loc[slovenia_df['medal_type'] == 'BRONZE', 'bronze_medal'] = slovenia_df['counts']
slovenia_df['bronze_medal'].fillna(0, inplace=True)

slovenia_df
```

```python
radio_group_color = pn.widgets.RadioBoxGroup(name='color combination', options=['Blue', 'Red', 'Black', 'Green', 'Yellow'], inline=True)
@pn.depends(radio_group_color.param.value)
def get_color(selected_option):
    if selected_option == 'Red':
        return 'red'
    elif selected_option == 'Blue':
        return 'blue'
    elif selected_option == 'Black':
        return 'black'
    elif selected_option == 'Green':
        return 'green'
    elif selected_option == 'Yellow':
        return 'yellow'
```

```python
#option=get_color
option = get_color(radio_group_color.value)
def set_color_map(option):
    if option == 'blue':
        return px.colors.sequential.Blues 
    elif option == 'red':
        return px.colors.sequential.Reds 
    elif option == 'black':
        return px.colors.sequential.Greys   
    elif option == 'green':
        return px.colors.sequential.Greens 
    elif option == 'yellow':
        return px.colors.sequential.Darkmint #Solar   
    
def create_map():
    option = get_color(radio_group_color.value)
    color_scale = set_color_map(option)
    fig= px.choropleth(
        worldmap,
        locations="country_name",
        locationmode='country names',
        color='sum_of_medals',
        hover_data='sum_of_medals', 
        animation_frame='date',
        title="Olympics Medal timelapse",
        color_continuous_scale=color_scale,
        width = 1500,
        height = 1200)
    return fig 
```

```python
country = list(df.country_name.unique())
country_widget = pn.widgets.Select(name="country_name", options=country)

def create_plot():
    idf = disc_dfx.interactive()
    idf = idf[(idf["country_name"] == country_widget)] #& (idf["discipline_title"] == discipline_widget)
    idf.head()
    second = idf.hvplot(kind="bar", x="discipline_title", y="counts", color = "medal_type", cmap=['blue', 'orange', 'red'])
    return second
```

```python
discipline = list(df2.discipline_title.unique())
discipline_widget = pn.widgets.Select(name="discipline_title", options=discipline)

def create_plot_date():
    idf = disc_dfx2.interactive()
    idf = idf[(idf["country_name"] == country_widget) & (idf["discipline_title"] == discipline_widget)]
    idf.head()
    third = idf.hvplot(kind="bar", x="date", y="counts", color = "medal_type", cmap=['blue', 'red', 'orange'])
    return third
```

```python
radio_group = pn.widgets.RadioBoxGroup(name='Olympic medals visualization', options=['By Country', 'By Discipline', 'Map'], inline=True)
```

```python
text = 'Choose visualization'
```

```python
text_color = 'Pick color combination for map'
text_color2 = 'Pick color combination for visualization by country'
```

```python
png_pane = pn.pane.PNG('https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcR_xnjnm4PXu1s264F_xK7pJv5q0yl5thNNRw&usqp=CAU')
```

```python
@pn.depends(radio_group.param.value) 
def update_plots(selected_option):
    if selected_option == 'By Country':
        return create_plot()
    elif selected_option == 'By Discipline':
        return create_plot_date()
    elif selected_option == 'Map':
        return create_map()
```

```python
# Instantiate the template with widgets displayed in the sidebar
template = pn.template.BootstrapTemplate(
    title='Olympics',
    sidebar=[text, radio_group, text_color, radio_group_color]
)

# Append the plots to the main area
template.main.append(png_pane)
template.main.append(update_plots)


template.servable()
```
