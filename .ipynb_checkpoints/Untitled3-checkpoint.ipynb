```python
radio_group_color = pn.widgets.RadioBoxGroup(name='color combination', options=['Blue', 'Red', 'Black', 'Green', 'Yellow'], inline=True)
@pn.depends(radio_group_color.param.value)
def get_color(selected_option):
    if selected_option == 'Red':
        return 'red'
    elif selected_option == 'Blue':
        return 'blue'
    elif selected_option == 'Black':
        return 'black'
    elif selected_option == 'Green':
        return 'green'
    elif selected_option == 'Yellow':
        return 'yellow'
```

```python
#option=get_color
option = get_color(radio_group_color.value)
def set_color_map(option):
    if option == 'blue':
        return px.colors.sequential.Blues 
    elif option == 'red':
        return px.colors.sequential.Reds 
    elif option == 'black':
        return px.colors.sequential.Greys   
    elif option == 'green':
        return px.colors.sequential.Greens 
    elif option == 'yellow':
        return px.colors.sequential.Darkmint #Solar   
    
def create_map():
    option = get_color(radio_group_color.value)
    color_scale = set_color_map(option)
    fig= px.choropleth(
        worldmap,
        locations="country_name",
        locationmode='country names',
        color='sum_of_medals',
        hover_data='sum_of_medals', 
        animation_frame='date',
        title="Olympics Medal timelapse",
        color_continuous_scale=color_scale,
        width = 1500,
        height = 1200)
    return fig 
```
